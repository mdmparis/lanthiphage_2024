# Retrieve the results from Antismash results

This part if how the results were retrieved from antismash genbank files.</br>
The data to run this code are not available here but the resulting table is saved in data/Proteins_lanthi-class-i.tsv</br>

Antismash_results_clean_lanthi_class-i.tsv is a table with all the genbank protocluster of lanthipeptide class-i detected in this study.</br>
This script was used for both retrieving all the protein inside the lanthipeptide as well as the lanthipeptide precursors (needed for the Script 03).

# Domain per system annotation

```python
Proteins=pd.read_table('./../Data/Proteins_lanthi-class-i.tsv',sep='\t')
synthetase=pd.read_table('./../Data/Results_01_Defense_and_MGE.tsv',sep='\t')

ClassI=Proteins.loc[Proteins.Type=='lanthipeptide-class-i']
ClassI['Id_prot']=ClassI.sys_id+'_'+ClassI.Accession
```

```python
# For proteins with multiple HMM hit, concatenate the profiles name in a Domains variable
Domains=(ClassI.groupby(['Id_prot','sys_id']).Domains.apply(lambda x: "|".join(x.sort_values()))).reset_index()
Domains.loc[Domains.sys_id.isin(synthetase.loc[synthetase.Type=='lanthipeptide-class-i'].Accession_DB.value_counts())]
```

For each system detected in antismash, calcul for each known domain of class I lanthipeptide:

LANC_like: Lanthionine synthetase</br>
Lant_dehydr_C|Lant_dehydr_N: Dehydratase with two domains.</br>
Lant_dehydr_N: Dehydratase without the "thiopeptide domain"</br>
PCMT: Methyltransferase</br>
Lant_dehydr_C: "Thiopeptide" domain</br>
Lant_dehydr_C|PCMT: Methyltransferase with the "Thiopeptide domain"</br>
Transporter: If a transporter is detected by antismash

```python
System_domain=pd.DataFrame({'sys_id':[],'Synthetase':[],'LanC':[],'Dehydra_N_Thiopep':[],'Dehydra':[],'Thiopep_Methyl':[],'Methyl':[],'Thiopep':[],'Transporter':[]})
System_domain=synthetase[['sys_id','Accession_DB']].merge(Domains[['sys_id']].drop_duplicates(),on="sys_id")
System_domain['LanC']=0
System_domain['Dehydra_N_Thiopep']=0
System_domain['Dehydra']=0
System_domain['Thiopep_Methyl']=0
System_domain['Methyl']=0
System_domain['Thiopep']=0
System_domain['Transporter']=0

for i in tqdm(System_domain.index.to_list()):
    tmp=Domains.loc[Domains.sys_id==System_domain.sys_id[i]]
    
    System_domain.loc[i,'LanC']=len(tmp.loc[(tmp.Domains.isin(['LANC_like','LANC_like|PCMT']))])
    System_domain.loc[i,'Dehydra_N_Thiopep']=len(tmp.loc[(tmp.Domains=='Lant_dehydr_C|Lant_dehydr_N')])
    System_domain.loc[i,'Dehydra']=len(tmp.loc[(tmp.Domains=='Lant_dehydr_N')])
    System_domain.loc[i,'Thiopep_Methyl']=len(tmp.loc[(tmp.Domains=='Lant_dehydr_C|PCMT')])
    System_domain.loc[i,'Methyl']=len(tmp.loc[(tmp.Domains=='PCMT')])
    System_domain.loc[i,'Thiopep']=len(tmp.loc[(tmp.Domains=='Lant_dehydr_C')])
    System_domain.loc[i,'Transporter']=len(tmp.loc[(tmp.Domains.str.contains('transport '))])
```

# Statistic on the different domains

```python
System_domain_info=System_domain.merge(synthetase[['Cluster','Accession_DB','sys_id','New_clade']],on=['sys_id','Accession_DB']).rename({'Accession_DB':'Synthetase'},axis=1)
System_domain_info['Defensive']='Non defensive'
System_domain_info.loc[~System_domain_info.New_clade.isin(['Non Defensive','Limit']),'Defensive']='Defensive'
System_domain_info
```

```python
System_domain_info['LanC_binary']=0
System_domain_info['Dehydra_N_Thiopep_binary']=0
System_domain_info['Dehydra_binary']=0
System_domain_info['Thiopep_Methyl_binary']=0
System_domain_info['Methyl_binary']=0
System_domain_info['Thiopep_binary']=0
System_domain_info['Transporter_binary']=0

System_domain_info.loc[System_domain_info.LanC>0,'LanC_binary']=1
System_domain_info.loc[System_domain_info.Dehydra_N_Thiopep>0,'Dehydra_N_Thiopep_binary']=1
System_domain_info.loc[System_domain_info.Dehydra>0,'Dehydra_binary']=1
System_domain_info.loc[System_domain_info.Thiopep>0,'Thiopep_Methyl_binary']=1
System_domain_info.loc[System_domain_info.Methyl>0,'Methyl_binary']=1
System_domain_info.loc[System_domain_info.Thiopep>0,'Thiopep_binary']=1
System_domain_info.loc[System_domain_info.Transporter>0,'Transporter_binary']=1
```

```python
import seaborn as sns
Matrix_domain_binary=System_domain_info[['Synthetase','LanC_binary','Dehydra_N_Thiopep_binary','Dehydra_binary','Thiopep_Methyl_binary','Methyl_binary','Thiopep_binary','Transporter_binary','New_clade']].sort_values('New_clade')
Colors=pd.DataFrame({'New_clade':['Non Defensive','Limit','Ohter','Clade 1','Clade 2'," Clade 3","Clade 4",'Clade 5','Clade 6','Clade 7','Clade 8','Clade 9','Clade 10'],
              'Colors':["black",'#bcbcbc','#eeeeee','#c3ad6f','#a40808','#e9d8a6',
                       '#b43b04','#ffb833','#de7002','#81180e','#7a6a3e','#641619','#491909']})
Matrix_domain_binary=Matrix_domain_binary.merge(Colors,on='New_clade')
sns.clustermap(Matrix_domain_binary.set_index('Synthetase').drop(columns=['New_clade','Colors']),yticklabels=False,row_cluster=False,
              cmap="GnBu",row_colors=Matrix_domain_binary.Colors.to_list())
```
